```python
import pandas as pd
```

```python
file = "../Resources/CountryPop.csv"
ofile = "../Resources/EconomicFreedom.csv"

countrydf = pd.read_csv(file)
ecodf = pd.read_csv(ofile)
```

```python
countrydf.head()
```

```python
ecodf.head()
```

```python
countrydf = countrydf[["Country","Population",
                       "Area (sq. mi.)","Pop. Density (per sq. mi.)"]]
countrydf.head()
```

```python
# Country had trailing spaces, this trims them
countrydf['Country'] = pd.core.strings.str_strip(countrydf['Country'])
```

```python
ecodf = ecodf[["year","countries","ECONOMIC FREEDOM",
              "rank","1d_top_marg_tax_rate",
                       "4c_black_market"]]
ecodf.head()
```

```python
countrydf.count()
```

```python
ecodf.count()
```

```python
ecodf.rename(columns={'countries':'Country','ECONOMIC FREEDOM':'Economic freedom', '1d_top_marg_tax_rate':'Top marginal tax rate', '4c_black_market': 'Black market'}, 
                 inplace=True)
```

```python
ecodf_copy = ecodf.copy()
```

```python
ecodf_copy.head()
ecodf_copy.count()
```

```python
# Drop all rows with missing information
ecodf_copy = ecodf_copy.dropna(subset=['year', 'Country', 'Economic freedom', 'rank','Top marginal tax rate'])
```

```python
ecodf_copy.count()
```

```python
ecodf_copy.head()
```

```python
df_merge = pd.merge(countrydf, ecodf_copy, on='Country')
```
