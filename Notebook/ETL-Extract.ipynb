# Load data into the database server.

```python
import sys, psycopg2
sys.path.append("..")
from config.pg_config import connection_cfg as conn_cfg
```

```python
from sqlalchemy import create_engine, Table, Column, MetaData
from sqlalchemy.orm import mapper, create_session
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.exc import IntegrityError
```

```python
from sqlalchemy import create_engine
db_string = f"postgres://{conn_cfg['user']}:{conn_cfg['password']}@{conn_cfg['host']}:{conn_cfg['port']}/{conn_cfg['database']}"
```

```python
print(f"postgres://{conn_cfg['user']}:<PASSWORD_HIDDEN>@{conn_cfg['host']}:{conn_cfg['port']}/{conn_cfg['database']}")
```

```python
try:
    db=create_engine(db_string)
    print("Successfully connected to Postgres database")
    session=create_session(bind=db)
except Exception as e:
    print(f"Exception occured while connecting to database of type {type(e).__name__}, {e}")
```

```python
try:
    db.execute("""DROP TABLE IF EXISTS COUNTRY_ECONOMIC_DATA""")
    db.execute("""DROP TABLE IF EXISTS COUNTRIES""")
    
    db.execute("""CREATE TABLE COUNTRIES (
            ID                              SERIAL             PRIMARY KEY,
            COUNTRY                         VARCHAR(100)       UNIQUE,
            POPULATION                      INTEGER, 
            AREA                            INTEGER,
            POPULATION_DENSITY              FLOAT
    )
    """)
    print("COUNTRIES table successfully created")
    
    
    db.execute("""CREATE TABLE COUNTRY_ECONOMIC_DATA (
            ID                              SERIAL             PRIMARY KEY,
            COUNTRY_ID                      INTEGER            REFERENCES COUNTRIES(ID),
            ECON_FREEDOM_INDEX              FLOAT, 
            RANK                            INTEGER,
            TOP_TAX_RATE                    FLOAT,
            BLACK_MARKET                    FLOAT
            
    )
    """)
    print("COUNTRY_ECONOMIC_DATA table successfully created")
except Exception as e:
        print(f"Exception occured while creating COUNTRY_ECONOMIC_DATA table of type {type(e).__name__}, {e.pgcode}")
```

```python
countries_records_inserted = 0
countries_failed_inserts = 0
countries_total_records = 0
for _, row in countrydf.iterrows():
    country_name = row["Country"]
    population = int(row["Population"])
    area = int(row["Area (sq. mi.)"])
    pop_density= float(row["Pop. Density (per sq. mi.)"].replace(",","."))

    try:
        countries_total_records += 1
        db.execute(f"INSERT INTO COUNTRIES(COUNTRY, POPULATION, AREA, POPULATION_DENSITY) VALUES ($${country_name}$$, {population}, {area}, {pop_density})")
        countries_records_inserted += 1
    except IntegrityError:
        print(f"Duplicate Insert for country {country_name}")
        countries_failed_inserts += 1
    except Exception as e:
        print(f"INSERT FAILED for {country_name}. {e}")
        countries_failed_inserts += 1

db.execute("COMMIT")
```

```python
header_width = 80    
print("~" * header_width)
print("Data Loading Statistics for COUNTRIES table")
print("~" * header_width)
print(f"Total Records Processed => {countries_total_records}\nTotal Records Inserted => {countries_records_inserted} \nFailed Inserts => {countries_failed_inserts}")
print("~" * header_width)
```

```python
econ_records_inserted = 0
econ_failed_inserts = 0
econ_total_records = 0
for _, row in df_merge.iterrows():
    country_name = row["Country"]
    year = int(row["year"])
    econ_freedom = float(row["Economic freedom"])
    rank = float(row["rank"])
    tax_rate = float(row["Top marginal tax rate"])
    black_market = float(row["Black market"])
    
    
    try:
        econ_total_records += 1
        db.execute(f"INSERT INTO COUNTRY_ECONOMIC_DATA(COUNTRY_ID, ECON_FREEDOM_INDEX, RANK, TOP_TAX_RATE, BLACK_MARKET) VALUES ((select id from countries where COUNTRY = $${country_name}$$), {econ_freedom}, {rank}, {tax_rate}, {black_market})")
        econ_records_inserted += 1
    except IntegrityError:
        print(f"Duplicate Insert for country {country_name}")
        econ_failed_inserts += 1
    except Exception as e:
        print(f"INSERT FAILED for {country_name}. {e}")
        econ_failed_inserts += 1

db.execute("COMMIT")
```

```python
header_width = 80    
print("~" * header_width)
print("Data Loading Statistics for COUNTRY_ECONOMIC_DATA table")
print("~" * header_width)
print(f"Total Records Processed => {econ_total_records}\nTotal Records Inserted => {econ_records_inserted} \nFailed Inserts => {econ_failed_inserts}")
print("~" * header_width)
```
